```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 500 #número de pasos a realiazar
reps = 5 #número de trayectorias a realizar
```

```python
def simular_PAS(p,n,reps):
  # Input:
  # p: probabilidad de sumar 1
  # n: número de pasos total
  # reps: número de trayectorias

  rng = np.random.default_rng() #establece la variable que va a generar los números aleatorios
  pasos = rng.choice( #crea el la matriz de tamano reps x n con valores en 1, -1 con probabilidad p, 1-p.
    [1, -1],
    size=(reps,n),
    p=[p, 1-p]
  )
  trayectoria=np.cumsum(pasos, axis=1) #suma los valores a lo largo del eje 1, que son las filas
  return trayectoria
```

Diferentes trayectorias del PAS para distintos valores de p

```python
p=0.5 #establece la probabilidad del PAS
trayectoria= simular_PAS(p,n,reps) #llama la funcion de simular el pas

x = np.linspace(0, n-1,n) #genera un vector con los valores 0,1,...,n-1
for i in range(0,reps,1): #por cada repeticion genera una trayecotria
    plt.plot(x, trayectoria[i, :]) #trayectoria[i,:] devuelve la fila i-esima

#Retoques esteticos
plt.xlabel('n')
plt.ylabel(r'$X_n$')
plt.grid(True, alpha=0.4)
plt.title(f'{reps} trayectorias de un PAS con p={p}')
plt.show()
```

Visualización de las trayectorias con la esperanza y desviación típica teóricas para distintos valores de p.

```python
x = np.arange(n)  # vector con los valores 0,1,...,n-1
p=0.5
# Esperanza y desviación típica teoricas segun la formula del PAS
esperanza = x * (2*p - 1)
sigma = np.sqrt(4*p*(1-p)*x)


trayectoria = simular_PAS(p,n,reps) #genera las trayectorias

for i in range(reps):
    plt.plot(x, trayectoria[i, :]) #dibujamos las trayectorias


plt.plot(x, esperanza, 'k--', linewidth=2, label='Media teórica') #linea de la esperanza teorica

# Bandas de las desviaciones tipicas
plt.fill_between(x, esperanza - sigma, esperanza + sigma, #se dibujan las regiones sombreadas de una desviacion tipica centrada en la media
                 color='gray', alpha=0.3, label='±1σ') #establece una opacidad del 30% y pone el nombre que luego sale en la leyenda

plt.fill_between(x, esperanza - 2*sigma, esperanza + 2*sigma,
                 color='gray', alpha=0.15, label='±2σ')

#Retoques esteticos
plt.xlabel('n')
plt.ylabel(r'$X_n$')
plt.grid(True, alpha=0.4)
plt.title(f'{reps} trayectorias de un PAS con p={p}')
plt.legend()
plt.show()
```

Visualización de la media teórica y la media observada tras cierto número de repeticiones

```python
x = np.arange(n)  #vector con los valores 0,1,...,n-1
p=0.5
esperanza = x * (2*p - 1) #esperanza teorica
plt.plot(x, esperanza, color='red', label='Media teorica')
num_reps=[5,100,200,500,1000] #definimos que trayectorias queremos simular

for i in range (5): #recorre el vector num_reps
  trayectoria = simular_PAS(p, n, num_reps[i]) #en cada iteracion me simula tantas traytorias como indique el indice i
  media_empirica = np.mean(trayectoria, axis=0) #calcula el vector de medias a traves del eje 0, o sea, por columnas
  plt.plot(x, media_empirica, label=f'{num_reps[i]} trayectorias')

#Retoques esteticos
plt.xlabel('n')
plt.ylabel(r'$E[X_n]$')
plt.title(f'Media teorica y medias observadas para p={p}')
plt.legend()
plt.grid(True, alpha=0.4)
plt.show()
```

```python
var = (4*p*(1-p)*x) #varianza teorica
p=0.5
plt.plot(x, var, color='red', label='Varianza teórica') #dibujamos la varianza teorica

for i in range (5): #recorre el vector num_reps
  trayectoria = simular_PAS(p, n, num_reps[i]) #en cada iteracion simula tantas traytorias como indique el indice i
  media_empirica = np.var(trayectoria, axis=0) #calcula el vector de varianzas a traves del eje 0, o sea, por columnas
  plt.plot(x, media_empirica, label=f'{num_reps[i]} trayectorias')

#Retoques esteticos
plt.xlabel('n')
plt.ylabel(r'$Var(X_n)$')
plt.title(f'Varianza teorica y varianzas observadas para diferentes números de repeticiones (p={p})')
plt.legend()
plt.grid(True, alpha=0.4)
plt.show()
```
